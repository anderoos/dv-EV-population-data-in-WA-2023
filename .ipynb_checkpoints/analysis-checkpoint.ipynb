```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Did Federal tax credit have a profound impact on EV sales in Washington?
# Which car was was the most popular in each county?
# When did car sales spike?
    # Convert the % of electric cars in county
# Pipeline
# Tax credit ended 2022
# 
# Importing data
df_ev_pop_WA = pd.read_csv('Resources/Electric_Vehicle_Population_Data.csv')
df_ev_pop_county = pd.read_csv('Resources/Electric_Vehicle_Population_Size_History_By_County.csv')

# Subframe of county data with only Washington State
df_ev_pop_county_WA = df_ev_pop_county[df_ev_pop_county['State'] == 'WA']
# df_ev_pop_county_WA.dtypes
df_ev_pop_county_WA['Date'] = pd.to_datetime(df_ev_pop_county_WA['Date'])
df_ev_pop_county_WA.head()
```

```python
# Cleaning
# Removing irrelevant data in df_ev_pop_WA
df_ev_pop_WA = df_ev_pop_WA.drop(['VIN (1-10)', 'Postal Code', 'Clean Alternative Fuel Vehicle (CAFV) Eligibility', 
                   'Legislative District', 'DOL Vehicle ID', 'Vehicle Location', 'Electric Utility', 
                   '2020 Census Tract'], axis=1)
df_ev_pop_WA.head()
```

```python
# General 
lst_makers = df_ev_pop_WA['Make'].unique()
make_model_yr = df_ev_pop_WA.groupby(['Make', 'Model', 'Model Year'])['County'].count()
make_model_yr.head()
```

```python
df_ev_pop_county_WA_by_county = df_ev_pop_county_WA.groupby('County')[['Date', 'Percent Electric Vehicles']]
df_ev_pop_county_WA_by_county.head()
# fig, ax = plt.subplots()
```
